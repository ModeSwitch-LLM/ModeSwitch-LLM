## Result Summary and Final Mode Selection

### What these results show
Our benchmark sweep produced a clear pattern across latency, throughput, energy per token, memory usage, and output-quality similarity to the FP16 baseline. The main takeaway is that there is **no single universal winner** across all workload types. Instead, different modes are strongest in different situations, which supports the core project idea: a **simple controller that switches modes based on workload characteristics** is more justified than always using one fixed mode.

### Strong overall findings
- **FP16 baseline** is stable and serves as the main reference, but it is consistently slower and less energy-efficient than several optimized modes.
- **INT8 quantization** is the strongest all-around practical candidate. It gives large latency and throughput improvements over FP16 across all major workload groups while keeping memory overhead small and preserving quality better than more aggressive 4-bit modes.
- **Speculative decoding** is very strong for short interactive workloads. It improves latency and throughput substantially relative to FP16 while keeping output quality effectively unchanged in the tested cases.
- **Prefix caching** clearly helps on shared-prefix workloads and should be treated as a conditional mode rather than a universal one.
- **Chunked prefill** helps on long-prompt and memory-heavy workloads, especially when prefill cost is high.
- **More aggressive 4-bit modes** like AWQ and GPTQ are very fast, but they introduce larger quality degradation than INT8 and therefore are less safe as default controller choices unless quality constraints are relaxed.

### Mode-by-mode interpretation

#### 1. FP16 baseline
FP16 is the reference point. It is not the best mode in efficiency, but it remains important because it provides the quality anchor and a stable fallback. We should **keep FP16 as the baseline and safe fallback**, not as the main optimized mode. 

#### 2. INT8 quantization
INT8 is the most balanced choice in the current results. It consistently improves latency and throughput by about **1.7× to 1.95×** over FP16 across the main workloads, while also reducing energy per token to roughly **0.42× to 0.51×** of baseline. Its memory overhead relative to FP16 is small compared with other optimized modes. This makes INT8 the **best general-purpose controller candidate**. 

#### 3. Speculative decoding
Speculative decoding performs especially well on short interactive workloads. On short-prompt workloads, it gives around **2.14× to 2.16×** latency/throughput improvement over baseline with **no measured quality degradation** in the comparison table. This makes it a strong candidate for the “fast interactive request” branch of the controller. Its current downside is that it was not benchmarked across as many workload categories as INT8. 

#### 4. Prefix caching
Prefix caching is not meant to be universally enabled; it is useful when repeated shared prefixes exist. In the shared-prefix workloads, it improves latency and throughput by about **1.38× to 1.39×** over baseline with **no measured quality degradation**. This makes it an excellent conditional mode: if shared-prefix structure is detected, prefix caching should be preferred. 

#### 5. Chunked prefill
Chunked prefill is helpful on long-prompt and memory-pressure workloads. It gives around **1.32× to 1.43×** speedup over FP16 with **no measured quality degradation** in the tested long-prompt and memory-heavy settings. This makes it a strong controller candidate for requests with large input contexts. 

#### 6. AWQ 4-bit
AWQ improves speed and energy relative to FP16, but it shows noticeably larger quality degradation than INT8 across several workloads. It is not a bad mode, but based on current results it is harder to justify as a primary controller choice when INT8 gives a better quality-efficiency tradeoff. AWQ is better treated as an aggressive optimization mode rather than a default recommended mode. 

#### 7. GPTQ 4-bit
GPTQ is the fastest and most energy-efficient mode in many tested workloads. For example, it reaches around **2.57× to 2.91×** speedup over FP16 on short-prompt workloads and **2.59× to 2.64×** on longer ones. However, its measured quality degradation is also among the highest, often much worse than INT8 and speculative decoding. So GPTQ looks strong as a “maximum-efficiency, lower-quality-tolerance” mode, but risky as a general-purpose controller option. 

#### 8. KV-cache compression
KV-cache compression underperformed relative to the other candidates. Its speedup over baseline is small, sometimes close to negligible, while memory overhead and energy efficiency are not compelling enough to justify keeping it in the first controller version. Based on current data, this mode should be removed from the main set. 

#### 9. Continuous batching
Continuous batching shows extremely high throughput, but the current evaluation is mostly single-request/controller-oriented rather than many-request serving-oriented. In that context, it is not the cleanest fit for the first controller version. It may still be valuable as a traffic-aware serving extension, but it should not be part of the initial small mode set for the controller. 

#### 10. CUDA graphs
CUDA graphs improve over baseline, but they do not stand out enough relative to INT8, speculative decoding, prefix caching, or chunked prefill. Since we want a small and interpretable controller mode library, CUDA graphs can be dropped from the first version without losing the main story. 

#### 11. AWQ + chunked prefill hybrid
The AWQ + chunked prefill hybrid does work and gives meaningful gains over baseline, but it does not clearly dominate INT8 or the cleaner specialized modes. It also inherits the quality concerns of AWQ. So while it is a useful proof-of-concept hybrid, it should not be one of the final core controller modes unless we specifically want one hybrid example in the writeup. 

### Modes to keep for the controller
Based on the current benchmark evidence, the best final controller set is:

1. **FP16 baseline** — keep as the safe fallback and reference.
2. **INT8 quantization** — keep as the best all-around efficiency mode.
3. **Speculative decoding** — keep for short, interactive, latency-sensitive requests.
4. **Prefix caching** — keep for repeated shared-prefix workloads.
5. **Chunked prefill** — keep for long-prompt / prefill-heavy / memory-heavy workloads.

This set is small, interpretable, and covers distinct workload regimes instead of containing several overlapping modes.

### Modes to cut from the first controller version
The following should be cut from the first controller version:

- **KV-cache compression** — weak gains relative to alternatives.
- **CUDA graphs** — useful, but not distinctive enough for the first small mode library.
- **Continuous batching** — better framed as a traffic-aware serving mode than a per-request controller mode.
- **AWQ 4-bit** — decent efficiency, but worse quality tradeoff than INT8.
- **GPTQ 4-bit** — very fast, but quality degradation is high enough that it is risky for the first main controller.
- **AWQ + chunked prefill hybrid** — works, but does not clearly beat simpler choices strongly enough.

### Final ranking for controller selection
If we rank the modes specifically for the **first version of the controller**, the order is:

1. **INT8 quantization**
2. **Speculative decoding**
3. **Chunked prefill**
4. **Prefix caching**
5. **FP16 baseline**
6. **GPTQ 4-bit**
7. **AWQ 4-bit**
8. **AWQ + chunked prefill**
9. **CUDA graphs**
10. **Continuous batching**
11. **KV-cache compression**

This ranking is not simply “fastest to slowest.” It prioritizes **practical controller value**, meaning a combination of:
- clear speed/energy benefit,
- manageable quality degradation,
- interpretable workload specialization,
- and usefulness in a simple rule-based controller.

### Recommended controller logic
A simple final controller can be framed as:

- **If request is short and interactive:** use **speculative decoding**
- **If request is long / prefill-heavy / memory-heavy:** use **chunked prefill**
- **If shared prefix is detected:** use **prefix caching**
- **Otherwise:** use **INT8 quantization**
- **Fallback:** use **FP16 baseline**

### Final takeaway
The experiments support the main project hypothesis: different inference modes are best for different workload regimes, so a small workload-aware selector is more justified than a single fixed mode. INT8 appears to be the strongest default optimized mode, while speculative decoding, prefix caching, and chunked prefill provide specialized gains in situations where they match the workload structure.
